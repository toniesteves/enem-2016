```python
import math
import pandas as pd
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from lightgbm import LGBMRegressor

import seaborn as sns
sns.set_style("whitegrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv(r'../data/final-train.csv', sep=';',encoding="UTF8").drop(['Unnamed: 0'],axis=1)
test = pd.read_csv('../data/final-test.csv', sep=';',encoding="UTF8")
```

```python
features = [
    'NU_NOTA_CN',
    'NU_NOTA_CH',
    'NU_NOTA_LC',
    'NU_NOTA_REDACAO',
    'NU_NOTA_COMP1',
    'NU_NOTA_COMP2',
    'NU_NOTA_COMP3',
    'NU_NOTA_COMP4',
    'NU_NOTA_COMP5']
```

```python
y_train = train['NU_NOTA_MT']

X_train = train[features]
X_test = test[features]
```

```python
scaler = StandardScaler()  

X_train = scaler.fit_transform(X_train)  
X_test = scaler.transform(X_test)
```

```python
lgbm_mdl =LGBMRegressor(
    num_leaves=100,
    learning_rate=0.05,
    n_estimators=1000)
```

```python
lgbm_mdl.fit(X_train, y_train)
```

```python
y_pred = lgbm_mdl.predict(X_test)
```

```python
y_pred_train = lgbm_mdl.predict(X_train)
```

```python
r2 = metrics.r2_score(y_train, y_pred_train)
rmse = math.sqrt(metrics.mean_squared_error(y_train, y_pred_train))
mae = metrics.mean_absolute_error(y_train, y_pred_train)
```

```python
lgbm_scores = pd.DataFrame({'LightGBM Regressor' : [r2, rmse, mae]}, index=['R2', 'RMSE', 'MAE'])
lgbm_scores.to_csv('../results/lgbm_scores.csv', sep=';', index=False)
lgbm_scores
```
